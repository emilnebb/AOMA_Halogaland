# Strain Data Exploration

```python
import numpy as np
import dataloader_halogaland.dataloader as dl
import matplotlib.pyplot as plt
import os
```

```python
loader = dl.Dataloader(os.getcwd()+'/../Data/Halogaland_2022_04_22')
file = loader.read_file(loader.anodes[0])

strain = loader.load_strain(loader.strain_names[0], file)
```

```python
plt.plot(strain['timestamp'], strain['SG1'])
plt.grid()
plt.show()
```

```python
strain_fft = np.fft.fft(strain['SG1'])
freq = np.fft.fftfreq(strain['timestamp'].shape[-1])

plt.plot(freq, strain_fft.real)
plt.xlabel("f [Hz]")
plt.ylabel("acceleration amplitude")
plt.grid()
plt.show()
```
